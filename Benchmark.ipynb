### Analyse du catalog Physio

#### Lecture du fichier excel à partir du scrapping

```python
physio_df = pd.read_excel("catalog_physio_20181129.xlsx", encoding="utf-8")
```

#### Création de la colonne Marque

```python
marques = []
noms = physio_df['nom']
for nom in noms:
    # on extrait la chaine de caractère qui précéde le nom du produit
    marque = nom.split(" ")[0]
    marques.append(marque)

marques
physio_df['marque'] = marques
```

#### Répartition des marques du catalogue

```python
marques_stat_physio = physio_df['marque'].describe()
marques_stat_physio
```

#### Nombre de produits par marque

```python
physio_df['marque'].value_counts()
```

#### Filtre des produits Chattanooga

```python
chattanooga_filter_physio = physio_df[physio_df['marque']=="CHATTANOOGA"]
chattanooga_filter_physio
```
